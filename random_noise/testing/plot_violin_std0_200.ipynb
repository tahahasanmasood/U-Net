```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# File paths for the CSV files
csv_files = [
    'denoised_psnr_values_std20.csv',
    'denoised_psnr_values_std40.csv',
    'denoised_psnr_values_std60.csv',
    'denoised_psnr_values_std80.csv',
    'denoised_psnr_values_std100.csv',
    'denoised_psnr_values_std120.csv',
    'denoised_psnr_values_std140.csv',
    'denoised_psnr_values_std160.csv',
    'denoised_psnr_values_std180.csv',
    'denoised_psnr_values_std200.csv'
]

# Combine all CSVs into a single DataFrame
all_data = pd.DataFrame()
for file in csv_files:
    data = pd.read_csv(file)
    all_data = pd.concat([all_data, data], ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_csv_path = 'combined_psnr_data.csv'
all_data.to_csv(combined_csv_path, index=False)

print(f"Combined data saved to '{combined_csv_path}'")

# Plot the violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='std', y='psnr_denoised', data=all_data, palette="muted")
plt.title('Denoised PSNR Values vs Noise Standard Deviation')
plt.xlabel('Standard Deviation of Noise')
plt.ylabel('PSNR (dB)')
plt.show()
```
